```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Google-Playstore.csv")
# Hiển thị thông tin tổng quan về dữ liệu
print("Thông tin tổng quan về dữ liệu:")
print(df.info())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Phân tích mô tả
print("Phân tích mô tả:")
print(df.describe())
```

```python

# Vẽ biểu đồ phân phối của các biến số
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], kde=True, color='blue')
plt.title('Phân phối của Rating')
plt.xlabel('Rating')
plt.ylabel('Số lượng')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df['Installs'], kde=True, color='green')
plt.title('Phân phối của Installs')
plt.xlabel('Installs')
plt.ylabel('Số lượng')
plt.show()
```

```python
# Phân tích đa biến
print("Phân tích đa biến:")
# Biểu đồ scatter plot giữa Rating và Installs
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rating', y='Installs', data=df)
plt.title('Biểu đồ scatter plot giữa Rating và Installs')
plt.show()
```

```python
# Biểu đồ pair plot giữa các biến số
sns.pairplot(df)
plt.title('Biểu đồ pair plot giữa các biến số')
plt.show()
```

```python

# Tính ma trận tương quan giữa các biến số
correlation_matrix = df.corr()
print("Ma trận tương quan giữa các biến số:")
print(correlation_matrix)
```

```python


# Phân tích ảnh hưởng
# Phân tích ảnh hưởng của biến số Price đối với biến số Rating bằng scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Rating', data=df)
plt.title("Biểu đồ scatter plot giữa biến số Price và Rating")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()

# Sử dụng kiểm định t để kiểm tra sự khác biệt ý nghĩa giữa các nhóm
# Ví dụ: Kiểm tra sự khác biệt giữa Rating của các ứng dụng miễn phí và có phí
free_apps = df[df['Free'] == 'TRUE']['Rating']
paid_apps = df[df['Free'] == 'FALSE']['Rating']

from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(free_apps, paid_apps, equal_var=False)
print("Giá trị t-statistic:", t_stat)
print("Giá trị p-value:", p_value)
if p_value < 0.05:
    print("Có sự khác biệt ý nghĩa giữa Rating của các nhóm miễn phí và có phí.")
else:
    print("Không có sự khác biệt ý nghĩa giữa Rating của các nhóm miễn phí và có phí.")

# Tổng kết và rút ra kết luận
# Tổng kết các phát hiện và kết luận từ phân tích EDA của bạn
```
